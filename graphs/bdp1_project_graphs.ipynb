```python
import seaborn as sns; sns.set()
import matplotlib.pyplot as plt
import numpy as np
```

```python
time = np.linspace(0,365,1000) # days

worker_cost = 0.0928*24 # day
master_cost = 0.0464*24 # day
storage_cost = 0.1/31 # day
storage_space = 1000
num_workers_1 = 100
num_workers_2 = 2

cost = (num_workers_1*worker_cost*time+master_cost*time+storage_cost*storage_space*time)/1000
```

```python
sns.lineplot(time, cost)
plt.xlabel("Time (days)")
plt.ylabel("Cost (k$)")
plt.savefig("../report/images/cost_graph.pdf")
```

```python
cpu_time = 151609100/60/60/24 #days

patients_1 = time/(cpu_time/(2*num_workers_1))
patients_2 = time/(cpu_time/(2*num_workers_2))
```

```python
sns.lineplot(time, patients_1)
sns.lineplot(time, patients_2)
plt.xlabel("Time (days)")
plt.ylabel("Number of Patients")
plt.savefig("../report/images/patient_graph.pdf")
```
